```python
import requests
url = "https://jsonplaceholder.typicode.com/posts"

response = requests.get(url)
if response.status_code == 200:
    try:
        posts = response.json()
        even_id_posts = [post for post in posts if post['id'] % 2 == 0]
        print("Посты с четными id:")
        for post in even_id_posts:
            print(post)
    except requests.exceptions.JSONDecodeError:
        print("Ошибка декодирования JSON.")
else:
    print(f"Ошибка запроса: {response.status_code}")
```

```python
import json
post_url = "https://jsonplaceholder.typicode.com/posts"
new_post_data = {
    "title": "Тестовый пост",
    "body": "Это содержимое тестового поста.",
    "userId": 1
}
post_response = requests.post(post_url, json=new_post_data)
if post_response.status_code == 201:
    try:
        created_post = post_response.json()
        print("Созданный пост:")
        print(json.dumps(created_post, indent=2))
    except requests.exceptions.JSONDecodeError:
        print("Ошибка декодирования JSON.")
else:
    print(f"Ошибка запроса: {post_response.status_code}")
```

```python
def update_post(post_id):
  url = f"https://jsonplaceholder.typicode.com/posts/{post_id}"

  updated_data = {
    "title": "Обновлённый пост"
    }

  response = requests.put(url, json=updated_data)
  print(response.json())

update_post(100)
```
